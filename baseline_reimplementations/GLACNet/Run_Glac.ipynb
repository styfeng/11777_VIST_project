```python
%cd /content/drive/MyDrive/Colab_Notebooks/11777/11777_VIST/11777_VIST_project/GLACNet/GLACNet
import json


test_json_path =  "/content/drive/MyDrive/Colab_Notebooks/11777/11777_VIST/11777_VIST_project/data_reformatting/stories_withcaptions_jsons/stories_test.json"
album_ids = []
with open(test_json_path) as f:
    test_data = json.load(f)
    f.close()
ids = list(test_data.keys())
for id in ids:
    album_id = test_data[id]['album_id']
    album_ids.append(album_id)


pred_json_path =  "/content/drive/MyDrive/Colab_Notebooks/11777/11777_VIST/11777_VIST_project/GLACNet/GLACNet/result.json"
with open(pred_json_path) as f:
    pred_data = json.load(f)
    f.close()

pred_stories = pred_data['output_stories']

result_dict = {}
for story in pred_stories:
    album_id = story['album_id']  
    if album_id in album_ids:
        result_dict[album_id] = story['story_text_normalized']

# Example sentence content for test result json file ["  there was a ton of food for the day . there was a lot of food to eat . the food was delicious . there was a lot of different types of food . all of the food was delicious . i had a great time ."]

output_path = "/content/drive/MyDrive/Colab_Notebooks/11777/11777_VIST/11777_VIST_project/GLACNet/GLACNet/prediction_test_GLACNet.json"
with open(output_path, "w") as json_file:
    #json_file.write(json.dumps(output, ensure_ascii=False))
    json_file.write(json.dumps(result_dict))
```
