# Quantitative Vis Analysis

```python

def check_obj_existence(story_fast_labels, story_content, image_captions):
    obj_count = 0
    result_containment = dict()
    result_notcontain = dict()
    for i in range(len(story_fast_labels)): # Loop through all images
        fast_labels = story_fast_labels[i]
        captions = image_captions[i] #corresponding caption
        story = story_content[i] #corresponding story
        
        for fl in fast_labels:
            # Check existence in caption for one image
            in_caption = False
            in_seq = False
            in_story = False
            for caption in captions:
                if fl in caption:
                    in_caption = True
            
            # Check existence in story for one image
            if fl in story:
                in_story = True
            
            if in_caption and in_story:
                both_count, caption_count, story_count, seq_count = result_containment.get(fl, (0,0,0,0))
                result_containment[fl] = (both_count + 1, caption_count, story_count, seq_count)
                
            elif in_caption:
                both_count, caption_count, story_count, seq_count = result_containment.get(fl, (0,0,0,0))
                result_containment[fl] = (both_count, caption_count + 1, story_count, seq_count)
                
            elif in_story:
                both_count, caption_count, story_count, seq_count = result_containment.get(fl, (0,0,0,0))
                result_containment[fl] = (both_count, caption_count, story_count + 1, seq_count)
                
            else:
                # Check all sequence for existence
                for story in story_content:
                    if fl in story:
                        in_seq = True
                        break
                if in_seq:
                    both_count, caption_count, story_count, seq_count = result_containment.get(fl, (0,0,0,0))
                    result_containment[fl] = (both_count, caption_count, story_count, seq_count + 1)
                else:
                    # Doesn't exist in anywhere in the story
                    result_notcontain[fl] = result_notcontain.get(fl, 0) + 1
    return result_containment, result_notcontain

def check_res_existence(story_res_labels, story_content, image_captions):
    obj_count = 0
    result_containment = dict()
    result_notcontain = dict()
    for i in range(len(story_res_labels)): # Loop through all images
        res_labels = story_res_labels[i]
        captions = image_captions[i] #corresponding caption
        story = story_content[i] #corresponding story
        
        for rl in res_labels:
            # Check existence in caption for one image
            in_caption = False
            in_story = False
            in_seq = False
            for caption in captions:
                if rl in caption:
                    in_caption = True
            
            # Check existence in story for one image
            if rl in story:
                in_story = True
            
            if in_caption and in_story:
                both_count, caption_count, story_count, seq_count = result_containment.get(rl, (0,0,0,0))
                result_containment[rl] = (both_count + 1, caption_count, story_count, seq_count)
                
            elif in_caption:
                both_count, caption_count, story_count, seq_count = result_containment.get(rl, (0,0,0,0))
                result_containment[rl] = (both_count, caption_count + 1, story_count, seq_count)
                
            elif in_story:
                both_count, caption_count, story_count, seq_count = result_containment.get(rl, (0,0,0,0))
                result_containment[rl] = (both_count, caption_count, story_count + 1, seq_count)
                
            else:
                # Check all sequence for existence
                for story in story_content:
                    if rl in story:
                        in_seq = True
                if in_seq:
                    both_count, caption_count, story_count, seq_count = result_containment.get(rl, (0,0,0,0))
                    result_containment[rl] = (both_count, caption_count, story_count, seq_count + 1)
                else:
                    # Doesn't exist in anywhere in the story
                    result_notcontain[rl] = result_notcontain.get(rl, 0) + 1

    return result_containment, result_notcontain



# resnet module gives image classes - get name of images
# - check if names of images are present in stories or not 
# solve them based on which story do not include any of the objects detected


# See what objects are common in both captiosn & stories 
# Important objects vs. non-important objects

# Present the TopK objects detected but not included in the story
```

```python
from PIL import Image, ImageDraw
from torchvision import transforms
from torchvision import utils as vutils
import matplotlib.pyplot as plt
import os
import torch
from coco_names import COCO_INSTANCE_CATEGORY_NAMES as coco_names
from imagenet1000_clsidx_to_labels import IMAGNET_CATEGORY_NAMES as imgnet_labels
import cv2
import numpy as np
from tqdm import tqdm

%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 10)
device = 'cpu'
if torch.cuda.is_available():
    device = 'cuda'

image_root_dir = '../../data/images/train'
font_path = "./fonts/arial.ttf"
save_path = "./quantitative"

resnet_preprocess = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

faster_rcnn_preprocess = transforms.Compose([
    transforms.ToTensor(),
])

img_classifier = img_classifier.to(device)
obj_detector = obj_detector.to(device)
# np.load(img_net_label_path, allow_pickle=True)

sum_number_objs = 0
num_images = 0
save_interval = 100

# Statistics
fast_obj_both_caption_story = dict()
fast_obj_caption_no_story = dict()
fast_obj_no_caption_story = dict()
fast_obj_no_caption_no_story = dict() # Mostly all going to be miss-classifications - Hard to distinguish btw 
fast_obj_no_story_in_seq = dict()

res_obj_both_caption_story = dict()
res_obj_caption_no_story = dict()
res_obj_no_caption_story = dict()
res_obj_no_caption_no_story = dict() # Mostly all going to be miss-classifications - Hard to distinguish btw 
res_obj_no_story_in_seq = dict()

# whether it was just mis-classification / or if it was intentionally omitted during story generation


threshold = 0.9
num_stories = 50
subset = 10
obj_detector = obj_detector.to(device)
img_classifier = img_classifier.to(device)

with torch.no_grad():
    for iteration, story_id in enumerate(tqdm(list(sis.Stories.keys()))):
        img_ids = sis.Stories[story_id]['img_ids']
        story_content = sis.Stories[story_id]['sent_texts']
        image_captions = sis.Stories[story_id]['captions']
        num_images += len(img_ids)
        story_fast_labels = []
        story_res_labels = []
        for i, img_id in enumerate(img_ids):
            obj_count = 0
            img_path = os.path.join(image_root_dir, img_id + ".jpg")
            img = Image.open(img_path)
            img = img.convert("RGB")
            
            # Run FASTGAN
            img_fast = faster_rcnn_preprocess(img)
            img_res = resnet_preprocess(img)
            img_fast = img_fast.unsqueeze(0).to(device)
            img_res = img_res.unsqueeze(0).to(device)

            preds = obj_detector(img_fast)[0]
            labels = preds["labels"].cpu().detach().numpy()
            scores = preds["scores"].cpu().detach().numpy()
            boxes = preds["boxes"].cpu().detach().numpy()
            
            # Run RESNET-152
            out = img_classifier(img_res)[0]
            label_i = torch.argmax(out)
            res_labels = imgnet_labels[label_i.item()].split(',')
            
            max_index = 0
            for score in scores:
                if(score < threshold):
                    break
                max_index += 1
            sum_number_objs += max_index
            
            fast_labels = []
            for l in set(labels[:max_index]):
                fast_labels.append(coco_names[l])
            story_fast_labels.append(fast_labels)
            story_res_labels.append(res_labels)

        fast_result_containment, fast_not_contain = check_obj_existence(story_fast_labels, story_content, image_captions)
        res_result_containment, res_not_contain = check_res_existence(story_res_labels, story_content, image_captions)
        if fast_result_containment:
            # update dictionaries
            for key in fast_result_containment.keys():
                both_count, caption_count, story_count, seq_count = fast_result_containment[key]
                if both_count != 0:
                    fast_obj_both_caption_story[key] = fast_obj_both_caption_story.get(key, 0) + both_count
                if caption_count != 0:
                    fast_obj_caption_no_story[key] = fast_obj_caption_no_story.get(key, 0) + caption_count
                if story_count != 0:
                    fast_obj_no_caption_story[key] = fast_obj_no_caption_story.get(key, 0) + story_count
                if seq_count != 0:
                    fast_obj_no_story_in_seq[key] = fast_obj_no_story_in_seq.get(key, 0) + seq_count
            
            for key in fast_not_contain.keys():
                fast_obj_no_caption_no_story[key] = fast_obj_no_caption_no_story.get(key, 0) + fast_not_contain[key]
                
        if res_result_containment:
            # update dictionaries
            for key in res_result_containment.keys():
                both_count, caption_count, story_count, seq_count = res_result_containment[key]
                if both_count != 0:
                    res_obj_both_caption_story[key] = res_obj_both_caption_story.get(key, 0) + both_count
                if caption_count != 0:
                    res_obj_caption_no_story[key] = res_obj_caption_no_story.get(key, 0) + caption_count
                if story_count != 0:
                    res_obj_no_caption_story[key] = res_obj_no_caption_story.get(key, 0) + story_count
                if seq_count != 0:
                    res_obj_no_story_in_seq[key] = res_obj_no_story_in_seq.get(key, 0) + seq_count
            
            for key in res_not_contain.keys():
                res_obj_no_caption_no_story[key] = res_obj_no_caption_no_story.get(key, 0) + res_not_contain[key]
    
       
        if((iteration + 1) % save_interval == 0):
            # Save dictionaries  
            np.save(save_path + "/fast_story_capt{}.npy".format(iteration), fast_obj_both_caption_story)
            np.save(save_path + "/fast_no_story_capt{}.npy".format(iteration), fast_obj_caption_no_story)
            np.save(save_path + "/fast_story_no_capt{}.npy".format(iteration), fast_obj_no_caption_story)
            np.save(save_path + "/fast_no_story_no_capt{}.npy".format(iteration), fast_obj_no_caption_no_story)
            np.save(save_path + "/fast_in_seq{}.npy".format(iteration), fast_obj_no_story_in_seq)

            np.save(save_path + "/res_story_capt{}.npy".format(iteration), res_obj_both_caption_story)
            np.save(save_path + "/res_no_story_capt{}.npy".format(iteration), res_obj_caption_no_story)
            np.save(save_path + "/res_story_no_capt{}.npy".format(iteration), res_obj_no_caption_story)
            np.save(save_path + "/res_no_story_no_capt{}.npy".format(iteration), res_obj_no_caption_no_story)
            np.save(save_path + "/res_in_seq{}.npy".format(iteration), res_obj_no_story_in_seq)

            
```

# Qualitative Vis Analysis

```python
# Qualitative Analysis 
# Look through 100 image story pairs and see if there are wrongly / undetected objects BUT present in story
from PIL import Image, ImageDraw, ImageFont
from torchvision import transforms
from torchvision import utils as vutils
import matplotlib.pyplot as plt
import os
import torch
from coco_names import COCO_INSTANCE_CATEGORY_NAMES as coco_names
import cv2
import numpy as np
from tqdm import tqdm

%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 10)
device = 'cpu'
if torch.cuda.is_available():
    device = 'cuda'

image_root_dir = '../../data/images/train'
font_path = "./fonts/arial.ttf"
save_path = "./qualitative"

faster_rcnn_preprocess = transforms.Compose([
    transforms.ToTensor(),
])

threshold = 0.75
num_stories = 100
start_story = 1000
obj_detector = obj_detector.to(device)
fnt = ImageFont.truetype(font_path, 50)

with torch.no_grad():
    for story_id in tqdm(list(sis.Stories.keys())[start_story:start_story + num_stories]):
        img_ids = sis.Stories[story_id]['img_ids']
        story_content = sis.Stories[story_id]['sent_texts']
#         sis.show_story(story_id)
        
        for i, img_id in enumerate(img_ids):
            img_path = os.path.join(image_root_dir, img_id + ".jpg")
            img = Image.open(img_path)
            width, height = img.size
            collage = Image.new('RGB',(2 * width, height + 100), (250,250,250))
            
            img = faster_rcnn_preprocess(img)
            img = img.unsqueeze(0).to(device)

            preds = obj_detector(img)[0]
            labels = preds["labels"].cpu().detach().numpy()
            scores = preds["scores"].cpu().detach().numpy()
            boxes = preds["boxes"].cpu().detach().numpy()
            max_index = 0
            
            for score in scores:
                if(score < threshold):
                    break
                max_index += 1

            img = Image.open(img_path)
            collage.paste(img, (0,0))


            for j, box in enumerate(boxes[:max_index]):
                d = ImageDraw.Draw(img)
                d.rectangle(box, outline = "Green", width = 10)
                d.text(box[:2], coco_names[labels[j]], fill = "Red", font = fnt)
            collage.paste(img, (width,0))
            im_write = ImageDraw.Draw(collage)
            im_write.text((0, height),story_content[i], fill = "Black", font = fnt)
            collage.save(save_path + "/{}_{}".format(story_id, i) + ".jpg")

```
